# Data [Visualization](https://pandas.pydata.org/docs/user_guide/visualization.html)
- Style Sheets
- Plot Types

Pandas has a built-in plotting library based off Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

For our plots in this notebook we will create series and dataframes of random numbers.

```python
df = pd.DataFrame(np.random.randn(50,10), columns=('one','two','three','four','five','six','seven','eight','nine','ten'))
```

```python
df.head()
```

## [Style Sheets](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

Matplotlib style sheets allow for consistent styling of your plots.

Here's a tutorial on [creating your own style sheets](https://www.youtube.com/watch?v=zpKhibvMJbM) if you're so inclined.


```python
df['one'].plot.hist()
```

### To add a style call `plt.style.use()`

```python
plt.style.use('classic')
df['one'].plot.hist()
```

```python
plt.style.use('dark_background')
df['one'].plot.hist()
```

```python
plt.style.use('ggplot')
df['one'].plot.hist()
```

```python
plt.style.use('grayscale')
df['one'].plot.hist()
```

Make sure to refer to the full [list of style sheets](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

## [Plot types](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)

There are two ways to call a plot, here are two examples with a histogram

`df['one'].plot.hist()`
`df['one'].plot(kind='hist')`

The kinds of plot to produce:

line : line plot (default)

bar : vertical bar plot

barh : horizontal bar plot

hist : histogram

box : boxplot

kde : Kernel Density Estimation plot

density : same as kde

area : area plot

pie : pie plot

scatter : scatter plot (DataFrame only)

hexbin : hexbin plot (DataFrame only)

```python
plt.style.use('seaborn-pastel') # We'll stick with a single style throughout
```

### [line plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html)

```python
df.plot.line()
```

### [vertical bar plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html)

```python
df['one'].plot.bar()
```

You can create stacked bar plots with the `stacked` parameter

```python
df.plot.bar(stacked = True)
```

### [horizontal bar plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.barh.html)

```python
df['one'].plot.barh()
```

Create stacked bar plots with the `stacked` parameter

```python
df.plot.barh(stacked=True)
```

### [histogram](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.hist.html)

```python
df.plot.hist()
```

Change the number of bins with the `bins` parameter

```python
df.plot.hist(bins=3)
```

Create stacked histograms with the `stacked` parameter

```python
df.plot.hist(stacked=True)
```

### [boxplot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.box.html)

```python
df.plot.box()
```

### [Kernel Density Estimation plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.kde.html)

```python
df.plot.kde()
```

### [area plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.area.html)

```python
area_df = pd.DataFrame(np.random.rand(50,10), columns=('one','two','three','four','five','six','seven','eight','nine','ten'))
```

```python
area_df.plot.area()
```

### [pie plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.pie.html)

```python
pie_series = pd.Series(np.random.rand(10))
```

```python
pie_series.plot.pie()
```

### [scatter plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html)

```python
df.plot.scatter(x='one',y='two')
```

### [hexbin plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.hexbin.html)

```python
hex_df = pd.DataFrame(np.random.randn(1000,2), columns=('x','y'))
```

```python
hex_df.plot.hexbin(x='x',y='y')
```

Define the number of hexagons in the x-direction with the `gridsize` parameter

```python
hex_df.plot.hexbin(x='x',y='y', gridsize=15)
```

This notebook has focussed on calling basic types of plots. As Pandas' plotting library is built upon Matplotlib it is best to learn how to use Matplotlib as well as Seaborn.

Don't worry if the documentation doesn't make sense at first it will become clearer as you are exposed to more parameters in Matplotlib.
