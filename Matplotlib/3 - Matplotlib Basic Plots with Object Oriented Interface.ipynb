# MatplotLib Object-Oriented Interface
- create a figure object with `plt.figure()`
- add axes with `fig.add_axes()`
- plot data with `axes.plot()`
- label axes with `axes.set_xlabel()` and `axes.set_ylabel()`
- title axes with `axes.set_title()`
- create multiple plots by calling multiple axes

Unlike using pyplot the Object-Oriented Interface involves creating a figure object (blank canvas) and calling methods off that object

It might seem more complicated than using pyplot at first but will give you much more control over your plots

Start by importing our libraries and creating our data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(-10,10,20)
y = x**3
```

## We instantiate a figure object using [plt.figure()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html)

Think of a figure object as a blank canvas

```python
fig = plt.figure()
```

## Add axes with [.add_axes()](https://www.geeksforgeeks.org/matplotlib-figure-figure-add_axes-in-python/)

`fig.add_axes([left,bottom,width,height])` - pass in a list of dimensions for the axes, values must be between 0 and 1 as they represent percentages of the total canvas size

`left` -  percentage in from the left of the canvas

`bottom` - percentage up from the bottom of the canvas

`width` - percentage width of the canvas total

`height` - percentage height of the canvas total

Don't worry if this doesn't make sense right away it will become clearer when we create multiple subplots below

```python
fig = plt.figure()

axes = fig.add_axes([0.5,0.5,0.5,0.5])
```

## Plot your data with [axes.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html)

`axes.plot(x,y)`

```python
fig = plt.figure()

axes = fig.add_axes([0.5,0.5,0.9,0.9])

axes.plot(x,y)
```

## Label your axes with [axes.set_xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html) and [axes.set_ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html)

```python
fig = plt.figure()

axes = fig.add_axes([0.5,0.5,0.9,0.9])

axes.plot(x,y)
axes.set_xlabel('THIS IS THE X AXIS')
axes.set_ylabel('THIS IS THE Y AXIS')
```

## Title your plots with [axes.set_title()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html)

```python
fig = plt.figure()

axes = fig.add_axes([0.5,0.5,0.9,0.9])

axes.plot(x,y)
axes.set_xlabel('THIS IS THE X AXIS')
axes.set_ylabel('THIS IS THE Y AXIS')
axes.set_title('THIS IS THE TITLE')
```

## Create subplots by adding multiple axes

Try playing with the parameters of `axes2` below to gain intuition of how they work

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,0.9,0.9]) # 0% in from the left, 0% up from the bottoms, takes up 90% of the canvas width and 90% of the canvas height
axes2 = fig.add_axes([0.3,0.3,0.4,0.5]) # 30% in from the left, 30% up from the bottom, takes up 40% of the canvas width and 50% of the canvas height
```

All we need to do now is use the methods we have already learnt to create our plot

```python
fig = plt.figure()  # creates a figure object

axes1 = fig.add_axes([0,0,1,1]) # creates axes 0% in from the left, 0% up from the bottom, takes up 100% of the canvas width and 100% of the canvas height
axes2 = fig.add_axes([0.5,0.15,0.3,0.3]) # creates axes 50% in from the left, 15% up from the bottoms, takes up 30% of the canvas width and 30% of the canvas height

axes1.plot(x,y) # plot x against y
axes1.set_title('AXES 1')   # set title to 'AXES 1'
axes1.set_xlabel('1 x')     # set x axis label to '1 x'
axes1.set_ylabel('1 y')     # set y axis label to '1y'

axes2.plot(y,x) # plot y against x
axes2.set_title('AXES 2')   # set title to 'AXES 2'
axes2.set_xlabel('2 x')     # set x axis label to '2 x'
axes2.set_ylabel('2 y')     # set y axis label to '2 y'
```
