# Figure Size, Legends and Saving Plots
- change size and resolution of plots with `figsize` and `dpi` arguments
- add a legend with `.legend()`
- save figures with `.savefig()`


Start by importing our libraries and creating our data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(-10,10,20)
y = x**3
```

## We can edit the size and shape of our plots by passing the `figsize` and `dpi` arguments into `plt.figure()`

`figsize=(width,height)` in inches

`dpi=n` dots per inches (100 by default)

```python
fig1 = plt.figure(figsize=(6,4),dpi=50)

axes = fig1.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig2 = plt.figure(figsize=(3,2),dpi=100)

axes = fig2.add_axes([0,0,1,1])
axes.plot(x,y)
```

Notice how the above plots are roughly the same size, although `fig1` has a figsize twice that of `fig2`, `fig2` has twice the dots per inch

The arguments are the same for subplots

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,5))

axes[0].plot(x,y)

axes[1].plot(y,x)
```

## Label your plots by passing the `label` argument to `axes.plot()`

Add a legend to your plot by calling the [`.legend()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) method

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label='x against y')
axes.plot(x,x**2,label='x against x squared')

axes.legend()
```

## We can specify where the legend appears in our plots by passing the `loc` argument in `axes.legend()`

For a full list of location codes see the [documentation](https://matplotlib.org/stable/api/legend_api.html)

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label='x against y')
axes.plot(x,x**2,label='x against x squared')

axes.legend(loc=4)  # lower right
```

## We can save our figures using [.savefig()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html#)

`fig.savefig('filename.extension')`

```python
fig.savefig('my_figure.jpg')
```
