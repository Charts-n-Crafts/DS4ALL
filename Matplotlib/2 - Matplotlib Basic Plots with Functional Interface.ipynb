```python
import numpy as np
x = np.linspace(-10,10,20)
y = x**3
```

## Create plots using [plt.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

Functional method for creating plots

```python
plt.plot(x,y)
```

Create labels for the axes with [plt.xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) and [plt.ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html) passing in a string as your axis labels.

```python
plt.plot(x,y)
plt.xlabel('THIS IS THE X AXIS')
plt.ylabel('THIS IS THE Y AXIS')
```

Create a title for your plots with [plt.title()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html)

```python
plt.plot(x,y)
plt.xlabel('THIS IS THE X AXIS')
plt.ylabel('THIS IS THE Y AXIS')
plt.title('THIS IS THE TITLE')
```

## Creating multiplots using [plt.subplots()](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html)

`plt.subplot(rows,columns,plot_number)`

```python
plt.subplot(1,2,1)  # one row, two columns, plot 1
plt.plot(x,y)

plt.subplot(1,2,2) # one row, two columns, plot 2
plt.plot(y,x)
```

```python
plt.subplot(2,1,1) # two rows, two columns, plot 1
plt.plot(x,y)

plt.subplot(2,1,2) # two rows, two columns, plot 2
plt.plot(y,x)
```
