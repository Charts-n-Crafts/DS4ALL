# Basic Subplots with Object Oriented Interface
- create subplots with `fig,axes = plt.subplots()`
- edit subplots by accessing via their index
- remove overlap with `plt.tight_layout()`

 Start by importing our libraries and creating our data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(-10,10,20)
y = x**3
```

We can create sublots by calling [`plt.sublots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) on `fig,axes`. This calls the `fig.add_axes()` method in the background

`fig,axes = plt.subplots(nrows=n,ncols=n)`

`nrows`= number of rows
`ncols` = number of columns



```python
fig,axes = plt.subplots(nrows=1,ncols=2)   # create 1 by 2 subplot
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1) # create 2 by 1 subplot
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2) # create 2 by 2 subplot
```

By calling axes we can now see there are four subplots

```python
axes
```

We can access these subplots by their index and call the usual methods to edit our plot

```python
fig,axes = plt.subplots(nrows=2,ncols=2) # create 2 by 2 subplot

axes[0,0].plot(x,y) # row 0, column 0 (top left)
axes[0,0].set_title('Plot [0,0]')

axes[1,1].plot(y,x) # row 1, column 1 (bottom right)
axes[1,1].set_title('Plot [1,1]')
```

As you can see the subplots are now overlapping, we can adjust this using [`plt.tight_layout()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html)

```python
fig,axes = plt.subplots(nrows=2,ncols=2) # create 2 by 2 subplot

axes[0,0].plot(x,y) # row 0, column 0 (top left)
axes[0,0].set_title('Plot [0,0]')

axes[1,1].plot(y,x) # row 1, column 1 (bottom right)
axes[1,1].set_title('Plot [1,1]')

plt.tight_layout()
```
