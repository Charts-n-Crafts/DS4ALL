# Customising Lines and Markers
- Line colour
- Line width
- Line transparency
- Line style
- Markers

Start by importing our libraries and creating our data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(-10,10,20)
y = x**3
```

Customise the colour of your plots by passing in the `colour` argument into `axes.plot()`

You can pass in [colour names](https://matplotlib.org/cheatsheets/_images/cheatsheets-2.png)

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='red')  # set line colour to red
```

[RGB Hex Codes](https://www.rapidtables.com/convert/color/rgb-to-hex.html)

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#699DA8')  # set line colour to HEX #699DA8
```

[Colour codes](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='m')  # set line colour to magenta
```

Define the thickness of the line by passing in the `linewidth` arguments into `axes.plot()`

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='m', linewidth=6)  # set line colour to magenta, six times the thickness of default
```

Define the transparency of the line by passing in the `alpha` argument into `axes.plot()`

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='m', linewidth=6, alpha=0.1)  # set line colour to magenta, six times the thickness of default, 10% opacity
```

Define the style of the line by passing in the [`linestyle`](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) argument into `axes.plot()`

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='m', linewidth=6, alpha=0.1, linestyle='-.')  # set line colour to magenta, six times the thickness of default, 10% opacity, dash dot
```

Mark your data points by passing the [`marker`](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html) argument into `axes.plot()`

Adjust you markers size with the following parameters

`markersize`        - size of the marker
`markerfacecolor`   - color of the marker face
`markeredgewidth`   - width of marker edge
`markeredgecolor`   - color of marker edge


```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='m', linewidth=1, alpha=0.5, linestyle='dotted',
           marker='D', markersize=10, markerfacecolor='c', markeredgewidth=2, markeredgecolor='b' )  # set line colour to magenta, default line-width, 50% opacity, dotted line, diamond marker tent times default size, cyan marker, two times default edge width, blue edge
```

Don't worry if this seems complicated at first you don't need to remember all this. Just bear in mind that there is a high degree of customisability, Your favourite search engine will help you easily find colour codes etc!
