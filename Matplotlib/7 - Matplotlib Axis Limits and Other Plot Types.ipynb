# Axis Limits and Other Plot types
- Set limits of x-axis with `.set_xlim()`
- Set limits of y-axis with `.set_ylim()`
- Create scatter plots with `plt.scatter()`
- Create histogram plots with `plt.hist()`
- Create boxplots with `plt.boxplot()`

Start by importing our libraries and creating our data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(-10,10,20)
y = x**3
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

## Set limits of x-axis with [`.set_xlim()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlim.html)

Pass in list with lower and upper bounds

`axes.set_xlim([lower, upper])`

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)

axes.set_xlim([-5,5]) # set lower bound of x-axis to -5 and upper bound to 5
```

## Set limits of y-axis with [`.set_ylim()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylim.html)

Pass in list with lower and upper bounds

`axes.set_ylim([lower, upper])`

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)

axes.set_ylim([-500,500]) # set lower bound of y-axis to -500 and upper bound to 500
```

## Matplotlib has many [plot types](https://matplotlib.org/stable/plot_types/index.html), however, for different plots we will likely want to use an extension of Matplotlib called [Seaborn](https://seaborn.pydata.org/)

Create scatter plots with [`plt.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
plt.scatter(x,y)
```

Create histogram plots with [`plt.hist()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
plt.hist(np.random.randn(303))  # a histogram of 303 random numbers from a standard normal distribution
```

Create box plots with [`plt.boxplot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html)

```python
plt.boxplot(x, data=y)
```
