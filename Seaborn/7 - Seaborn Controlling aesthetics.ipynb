# Controlling aesthetics
- Style
- Size and aspect
- Scale and context
- Palettes and colours

Seaborn plots are highly customizable so this notebook gives a high level overview of some of the options, make sure to check the documentation for:
- [Theming](https://seaborn.pydata.org/api.html#themeing)
- [Colour palettes](https://seaborn.pydata.org/api.html#color-palettes)
- [Palette widgets](https://seaborn.pydata.org/api.html#palette-widgets)
- [Utility functions](https://seaborn.pydata.org/api.html#utility-functions)

Seaborn [controlling figure aesthetics tutorial](https://seaborn.pydata.org/tutorial/aesthetics.html)

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = sns.load_dataset('penguins')
data.head()
```

## Control the general style of a plot with [`set_style`](https://seaborn.pydata.org/generated/seaborn.set_style.html#seaborn.set_style)

Style options: darkgrid, whitegrid, dark, white, ticks

```python
sns.set_style('darkgrid')
sns.countplot(x='species',data=data)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='species',data=data)
```

```python
sns.set_style('dark')
sns.countplot(x='species',data=data)
```

```python
sns.set_style('white')
sns.countplot(x='species',data=data)
```

```python
sns.set_style('ticks')
sns.countplot(x='species',data=data)
```

Remove the top and right spines from plots with [`.despine`](https://seaborn.pydata.org/generated/seaborn.despine.html#seaborn.despine)

```python
sns.set_style('ticks')
sns.countplot(x='species',data=data)
sns.despine()
```

`.despine()` removes the top and right spines by default. You can also remove the other spines using the `left` and `bottom` parameters

```python
sns.set_style('ticks')
sns.countplot(x='species',data=data)
sns.despine(left=True,bottom=True)
```

## Size and aspect

As seaborn makes calls to Matplotlib 'under the hood' you can use Matplotlib commands to effect e.g. the size and aspect ratio of your plots

```python
plt.figure(figsize=(12,6))
sns.countplot(x='species',data=data)
```

## Scale and context

using [`.set_context()`](https://seaborn.pydata.org/generated/seaborn.set_context.html#seaborn.set_context) you can change the scaling of various plot elements such as font size

context types: paper, notebook, talk, poster (notebook by default)

```python
sns.set_context('paper')
sns.countplot(x='species',data=data)
```

```python
sns.set_context('notebook')
sns.countplot(x='species',data=data)
```

```python
sns.set_context('talk')
sns.countplot(x='species',data=data)
```

```python
sns.set_context('poster')
sns.countplot(x='species',data=data)
```

You can change the scale of the font with the `font_scale` parameter

```python
sns.set_context('poster',font_scale=0.5)
sns.countplot(x='species',data=data)
```

## Palettes and colours

[Matplotlib colour maps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

Seaborn tutorial on [choosing colour palettes](https://seaborn.pydata.org/tutorial/color_palettes.html)

```python
sns.pairplot(data, hue='species', palette='coolwarm')
plt.savefig('pairplotwithpalette.png')
```
