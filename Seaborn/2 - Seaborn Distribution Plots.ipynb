# [Distribution plots](https://seaborn.pydata.org/api.html#distribution-plots)
- Visualize distributions with displot
- Editing displot with the `bins` and `kde` parameters
- Visualize joint distributions with jointplot
- Plot pairwise relationships across a dataframe with pairplot
- Colouring data by category using the `hue` parameter

Start by importing our libraries, by convention seaborn is imported as sns

```python
import seaborn as sns
%matplotlib inline
```

## Seaborn comes with multiple [built-in datasets](https://github.com/mwaskom/seaborn-data)

We can load in seaborn datasets using [`sns.load_dataset()`](https://seaborn.pydata.org/generated/seaborn.load_dataset.html)

```python
data = sns.load_dataset('penguins')
```

```python
data.head()
```

```python
data.describe()
```

## [displot](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot)

Visualise the data distribution of a variable by passing in a single column of your dataframe to `sns.displot()`

```python
sns.displot(data['flipper_length_mm'])
```

We can set the number of bins and add a [kernel density estimate](https://en.wikipedia.org/wiki/Kernel_density_estimation) with the following parameters

```python
sns.displot(data['flipper_length_mm'], bins=30, kde=True)
```

## [jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

Two joint displots for bivariate data, pass in x, pass in y, pass in dataset

Using the `kind` parameter you can choose between different plot types

`kind{ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }`

```python
sns.jointplot(x='bill_length_mm',y='flipper_length_mm',data=data)   # Scatter plot by default
```

```python
sns.jointplot(x='bill_length_mm',y='flipper_length_mm',data=data,kind='hex')    # Hex plot
```

```python
sns.jointplot(x='bill_length_mm',y='flipper_length_mm',data=data,kind='reg')    # Linear Regression Fit
```

```python
sns.jointplot(x='bill_length_mm',y='flipper_length_mm',data=data,kind='kde') # Kernel Density Estimation
```

## [pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)
Plot pairwise relationships across a dataframe, this is an especially useful plot for exploratory data analysis.

```python
sns.pairplot(data)
```

We can pass the name of a categorical variable into the `hue` parameter to colour the datapoints based on their category

```python
sns.pairplot(data,hue='species')
```

## [rugplot](https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot)

Plot marginal distributions with ticks along the x and y axes

```python
sns.rugplot(data['body_mass_g'])
```

This is just a basic introduction to using some of Seaborn's distribution plots, hopefully you can see how powerful this library is in allowing you to create visualisations with very little code, for more plots and how to modify them check the [api documentation](https://seaborn.pydata.org/api.html#distribution-plots), and [official tutorials](https://seaborn.pydata.org/tutorial/distributions.html)
