# Seaborn

Seaborn is a Python data visualization library building on top of Matplotlib. It is a [high-level interface](https://joss.theoj.org/papers/10.21105/joss.03021), that is more comfortable handling Pandas data frames and easier to make [great looking plots](http://seaborn.pydata.org/examples/index.html).

[Seaborn github](https://github.com/mwaskom/seaborn)

[Seaborn Official Documentation](https://seaborn.pydata.org/)

## Installation

Activate your conda environment with `conda activate <env_name>`

Install the seaborn library with

`conda install seaborn`

If you're not using Anaconda, heres a link to [other installation methods](https://seaborn.pydata.org/installing.html)
