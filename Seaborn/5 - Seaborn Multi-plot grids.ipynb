# [Multi-plot grids](https://seaborn.pydata.org/api.html#multi-plot-grids)
- Plot pairwise relationships in a dataset with `pairplot`
- Create subplot grids for pairwise relationships with `PairGrid`
- Create Multi-plot grids for plotting conditional relationships with `FacetGrid`

```python
import seaborn as sns
%matplotlib inline
data = sns.load_dataset('penguins')
data.head()
```

## [pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot)

```python
sns.pairplot(data)
```

## [PairGrid](https://seaborn.pydata.org/generated/seaborn.PairGrid.html#seaborn.PairGrid)

Choose what plot types you want to map

[tutorial](https://seaborn.pydata.org/tutorial/axis_grids.html#grid-tutorial)

```python
sns.PairGrid(data)
```

Add plots to your grid by calling the `map()` method

```python
grid = sns.PairGrid(data)
grid.map(sns.scatterplot)
```

Define the plot type along the diagonal with `.map_diag()`

```python
grid = sns.PairGrid(data)
grid.map_diag(sns.histplot)
```

Define the plot type along the upper half of the grid with `.map_upper()`

```python
grid = sns.PairGrid(data)
grid.map_diag(sns.histplot)
grid.map_upper(sns.scatterplot)
```

Define the plot type along the lower half of the grid with `.map_lower()`

```python
grid = sns.PairGrid(data)
grid.map_diag(sns.histplot)
grid.map_upper(sns.scatterplot)
grid.map_lower(sns.kdeplot)
```

## [FacetGrid](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html#seaborn.FacetGrid)

```python
grid = sns.FacetGrid(data=data,col='sex',row='species')
```

Map plot types to your grid with `.map()`

```python
grid = sns.FacetGrid(data=data,col='sex',row='species')
grid.map(sns.histplot, 'flipper_length_mm')
```

```python
grid = sns.FacetGrid(data=data,col='sex',row='species')
grid.map(sns.scatterplot, 'bill_length_mm','bill_depth_mm')
```
