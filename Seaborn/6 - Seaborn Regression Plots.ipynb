# [Regression plots](https://seaborn.pydata.org/api.html#regression-plots)

```python
import seaborn as sns
%matplotlib inline
data = sns.load_dataset('penguins')
data.head()
```

## [lmplot](https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot)

Plot data and regression model

```python
sns.lmplot(x='flipper_length_mm', y='body_mass_g',data=data)
```

Groub by category with the `hue` parameter

```python
sns.lmplot(x='flipper_length_mm', y='body_mass_g',data=data,hue='species')
```

Seperate categories into different subplots with the `col` and `row` parameters

```python
sns.lmplot(x='flipper_length_mm', y='body_mass_g',data=data,col='species')
```

```python
sns.lmplot(x='flipper_length_mm', y='body_mass_g',data=data,col='species',row='sex')
```
