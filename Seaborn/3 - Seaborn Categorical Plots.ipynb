# [Categorical plots](https://seaborn.pydata.org/api.html#categorical-plots)
- Draw categorical plots with catplot
- Show point estimates and confidence intervals as rectangular bars with barplot
- Show the count of observations in each categorical bin using bars with countplot
- Draw a box plot to show distributions with respect to categories with boxplot
- Draw a combination of boxplot and kernel density estimate with violinplot
- Draw a scatterplot where one variable is categorical with stripplot
- Draw a categorical scatterplot with non-overlapping points with swarmplot
- Combine plot types



```python
import seaborn as sns
import numpy as np
%matplotlib inline
data = sns.load_dataset('penguins')
data.head()
```

## [catplot](https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot)

catplot allows you to create any kind of categorical plot using the `kind` parameter

Options are: “strip”, “swarm”, “box”, “violin”, “boxen”, “point”, “bar”, or “count”.

We will look at each individually below

```python
sns.catplot(x='species',y='flipper_length_mm',data=data,kind='point',hue='sex')
```

## [barplot](https://seaborn.pydata.org/generated/seaborn.barplot.html#seaborn.barplot)

Show point estimates and confidence intervals as rectangular bars. By default, the estimator is the mean.

```python
sns.barplot(x='sex',y='bill_length_mm',data=data)
```

Change the estimator using the `estimator` parameter

```python
sns.barplot(x='sex',y='bill_length_mm',data=data,estimator=np.std)
```

Like most plots you can use the `hue` parameter to group by category

```python
sns.barplot(x='sex',y='bill_length_mm',data=data,estimator=np.std,hue='species')
```

## [countplot](https://seaborn.pydata.org/generated/seaborn.countplot.html#seaborn.countplot)

Show the counts of observations in each categorical bin using bars.

This is the same as visualizing the [`pd.count()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.count.html) method of a series

```python
sns.countplot(x='species',data=data)
```

## [boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn.boxplot)

Visualise distributions with respect to categories.

```python
sns.boxplot(x='species',y='flipper_length_mm',data=data)
```

Like most plots you can use the `hue` parameter to group by category

```python
sns.boxplot(x='species',y='flipper_length_mm',data=data,hue='sex')
```

## [violin plot](https://seaborn.pydata.org/generated/seaborn.violinplot.html#seaborn.violinplot)

Visualise a combination of boxplot and kernel density estimate.

```python
sns.violinplot(x='species',y='flipper_length_mm',data=data)
```

Like most plots you can use the `hue` parameter to group by category, with violing plots you can also use the `split` parameter to split the halves of the violin between two categories, making it easier to directly compare the distributions

```python
sns.violinplot(x='species',y='flipper_length_mm',data=data,hue='sex',split=True)
```

## [stripplot](https://seaborn.pydata.org/generated/seaborn.stripplot.html#seaborn.stripplot)

Draw a scatterplot where one variable is categorical.

```python
sns.stripplot(x='species',y='flipper_length_mm',data=data)
```

If there is overlap between data points you can use the `jitter` parameter to make seeing the distribution easier

```python
sns.stripplot(x='species',y='flipper_length_mm',data=data,jitter=True,hue='sex')
```

Like using `split` with violinplots you can use the `dodge` and `hue` parameters to split distributions by category

```python
sns.stripplot(x='species',y='flipper_length_mm',data=data,jitter=True,hue='sex',dodge=True)
```

## [swarmplot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html#seaborn.swarmplot)

Draw a categorical scatterplot with non-overlapping points.

This is like combining a stripplot with a violinplot

As you can see these plots can struggle with large datasets

```python
sns.swarmplot(x='species',y='flipper_length_mm',data=data)
```

## Combining plots

You can combine plots by calling them in the same cell

```python
sns.violinplot(x='species',y='flipper_length_mm',data=data)
sns.swarmplot(x='species',y='flipper_length_mm',data=data,color='black')
```
