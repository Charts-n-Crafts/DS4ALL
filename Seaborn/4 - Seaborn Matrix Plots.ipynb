# [Matrix Plots](https://seaborn.pydata.org/api.html#matrix-plots)
- heatmap
- clustermap

```python
import seaborn as sns
%matplotlib inline
data = sns.load_dataset('penguins')
data.head()
```

Create pairwise correlation of columns with [`df.corr()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html)

```python
data_corr = data.corr()
data_corr.head()
```

## [heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn.heatmap)

```python
sns.heatmap(data_corr)
```

Annotate the values within the heatmap using the `annot` parameter

```python
sns.heatmap(data_corr,annot=True)
```

Color the lines that divide cells with `linecolor` and set the width of the lines with`linewidths`

```python
sns.heatmap(data_corr,annot=True,linecolor='black',linewidths=3)
```

Set the [colourmap](sns.heatmap(data_corr,annot=True,linecolor='black',linewidths=3)) with the `cmap` parameter

```python
sns.heatmap(data_corr,annot=True,linecolor='black',linewidths=3,cmap='coolwarm')
```

```python
data_pivot = data.pivot_table(index='sex', columns='species',values='body_mass_g')
data_pivot
```

```python
sns.heatmap(data_pivot)
```

## [cluster map](https://seaborn.pydata.org/generated/seaborn.clustermap.html#seaborn.clustermap)

```python
sns.clustermap(data_corr)
```

Set the [colourmap](sns.heatmap(data_corr,annot=True,linecolor='black',linewidths=3)) with the `cmap` parameter

```python
sns.clustermap(data_corr,cmap='viridis')
```

Standardise the dimension to 0-1 with the `standard_scale` parameter (0=rows, 1=columns)

```python
sns.clustermap(data_corr,cmap='viridis',standard_scale=0)
```

Heatmap will display things in the order they are presented, a clustermap will cluster things so that similar groups are close to each other
